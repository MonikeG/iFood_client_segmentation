### Matriz de correlação dos dados

```python
import statsmodels.api as sm
import seaborn as sns

correlacao = df.corr()
#plotar correlação
plt.figure(figsize = (20,15))
plot = sns.heatmap(correlacao, annot = True, fmt=".1f", linewidths=.6)
plot
```

```python
corr_triu = correlacao.where(~np.tril(np.ones(correlacao.shape)).astype(np.bool))
corr_triu = corr_triu.stack()
corr_triu.name = 'Pearson Correlation Coefficient'
corr_triu[corr_triu > 0.7].to_frame()
```

Através dos coeficientes de correlação é possível ver que, em geral, quem tem maior renda anual também são pessoas que gastam mais com vinhos carnes e tem um gasto total maior no comércio durante o ano e utilizam mais descontos para realizar sua compra.
Além desse grupo com maior renda anual, um segundo grupo que tem um maior gasto com produtos regulares, também tendem a utilizar mais descontos em suas compras.


```python
corr_triu_neg = correlacao.where(~np.tril(np.ones(correlacao.shape)).astype(np.bool))
corr_triu_neg = corr_triu_neg.stack()
corr_triu_neg.name = 'Pearson Correlation Negative Coefficient'
corr_triu_neg[corr_triu_neg <  -0.7].to_frame()
```
